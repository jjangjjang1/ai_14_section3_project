```python
def score(soup):
    model = [i.findNext("td") for i in soup.table.find_all("td","specs poslabel")]
    cpu = [i.findNext("td") for i in model]
    gpu = [i.findNext("td") for i in cpu]
    memory = [i.findNext("td") for i in gpu]
    size = [i.findNext("td") for i in memory]
    weight = [i.findNext("td") for i in size]
    r23 = [i.findNext("td").findNext("td") for i in weight]
    #return  td_text(i for i in(model,cpu,gpu,size,weight,r23))
    return [td_text(i) for i in (model,cpu,gpu,memory,size,weight,r23)]
```

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By as by
import time
```

```python
def shop_init(driver):
    path = '//*[@id="__next"]/div/div[2]/div[2]/div[3]/div[1]/div[2]/div/div[2]/div[4]/a[2]'
    driver.find_element(by.XPATH,path).click()
    time.sleep(0.5)
    menu = '//*[@id="__next"]/div/div[2]/div[2]/div[3]/div[1]/div[2]/div/div[2]/div[2]'
    driver.find_element(by.XPATH,menu).click()
    path = '//*[@id="__next"]/div/div[2]/div[2]/div[3]/div[1]/div[2]/div/div[2]/div[2]/ul/li[3]'
    menu = '//*[@id="__next"]/div/div[2]/div[2]/div[3]/div[1]/div[2]/div/div[2]/div[2]'
    driver.find_element(by.XPATH,path).click()
    time.sleep(0.5)
    driver.find_element(by.XPATH,menu).click()
    path = '//*[@id="__next"]/div/div[2]/div[2]/div[3]/div[1]/div[2]/div/div[2]/div[2]/ul/li[5]/a'
    menu = '//*[@id="__next"]/div/div[2]/div[2]/div[3]/div[1]/div[2]/div/div[2]/div[2]'
    driver.find_element(by.XPATH,path).click()
    time.sleep(0.5)
    driver.find_element(by.XPATH,menu).click()
    path = '//*[@id="__next"]/div/div[2]/div[2]/div[3]/div[1]/div[2]/div/div[2]/div[2]/ul/li[7]/a'
    driver.find_element(by.XPATH,path).click()
    path = '//*[@id="__next"]/div/div[2]/div[2]/div[3]/div[1]/div[2]/ul/li[2]/a'
    driver.find_element(by.XPATH,path).click()
    return
```

```python
def item_select(num1,num2):
    driver.switch_to.window(window_name=driver.window_handles[0])
    path = f'//*[@id="__next"]/div/div[2]/div[2]/div[3]/div[1]/div[3]/div/ul[{num1}]/li[{num2}]/div[1]/a'
    driver.find_element(by.XPATH,path).click()
    return
```

```python
def next_page():
    driver.switch_to.window(window_name=driver.window_handles[0])
    driver.find_element(by.XPATH,'//*[contains(text(), "다음")]').click()
    return
```

```python
def get_name_list():
    name_list = ["상품명","최저가","제조사","브랜드","평점","화면크기","무게","램","CPU","해상도"]
    tmp = [driver.find_elements(by.TAG_NAME,"th")[i].text for i in range (len(driver.find_elements(by.TAG_NAME,"th")))]
    return [name_list + ["인텔 GPU"]if "인텔 GPU" in tmp else name_list + ["NVIDIA GPU"] if "NVIDIA GPU" in tmp else name_list + ["외장그래픽"] if "외장그래픽" in tmp else name_list + ["AMD GPU"] if "AMD GPU" in tmp else name_list + ["CPU"]][0]
```

```python
def catch(func, handle=lambda e : e, *args, **kwargs):
    try:
        return func(*args, **kwargs)
    except Exception as e:
        return "null"
```

```python
def get_data():
    name_list = get_name_list()
    data = [catch(lambda:driver.find_element(by.XPATH,f"//*[contains(text(), '{i}')]/following-sibling::td").text) for i in name_list]
    driver.execute_script("window.scrollTo(0, 0)")
    ai_pick = [catch(lambda:driver.find_elements(by.CLASS_NAME,'bestReview_cate__Zgp_c')[i].text) for i in range(3)]
    print(ai_pick)
    return name_list,[data[0].split("\n")[1]]+data[1:] + [1 if "화면" in ai_pick else 0] +[1 if "무게" in ai_pick else 0] + [1 if "디자인" in ai_pick else 0] + [1 if "색상" in ai_pick else 0] + [1 if "소음" in ai_pick else 0] + [1 if "사용성" in ai_pick else 0] + [1 if "발열" in ai_pick else 0] 
```

```python
def tab_select():
    driver.switch_to.window(window_name=driver.window_handles[1])
    try:
        driver.find_element(by.LINK_TEXT,'AiTEMS추천').click()
        driver.find_element(by.XPATH,'//*[contains(text(), "제품정보 더보기")]').click()
    except:
        print("클릭오류")
        return 1
    return 0
```

```python
def scroll(i):
    return driver.execute_script(f"window.scrollTo(0, {600+650*(i-1)})")
```

```python
def scarping(wheel,row,col,df):
    driver.switch_to.window(window_name=driver.window_handles[0])
    scroll(wheel)
    time.sleep(1)
    item_select(row,col)
    time.sleep(0.05)
    error = tab_select()
    if error == 1:
        driver.close()
        return df
    time.sleep(0.05)
    name_list,data_list = get_data()
    name_list[-1] = 'GPU'
    name_list = name_list + ["화면추천","무게추천","디자인추천","색상추천","소음추천","사용성추천","발열추천"]
    df = df.append({l:i for l,i in zip(name_list,data_list)},ignore_index=True)
    driver.close()
    return df
```

```python
def start(df):
    driver.switch_to.window(window_name=driver.window_handles[0])
    for i in range (1,len(driver.find_elements(by.CLASS_NAME,"imgList_search_list__X1rdh"))+1):
        l = 1 if i==1 else 2 if i==3 else 3 if i==5 else 4 if i in [7,8] else 5
        for j in range (1,6):
            df = scarping(l,i,j,df)
    return df
```

```python
def eda(df):
    df_copy = df.copy()
    
    str_expr = "gpu.str.contains('Xe',case=False) or gpu.str.contains('UHD',case=False) or gpu.str.contains('hd',case=False) or gpu.str.contains('Iris',case=False)" 
    q_df = df_copy.query(str_expr)
    q_index = q_df.index
    df_copy.loc[q_index,"gpu"] = "integrated graphics"
    
    str_expr = "gpu.str.contains('rtx',case=False)" 
    q_df = df_copy.query(str_expr)
    q_index = q_df.index
    df_copy.loc[q_index,"gpu"] = ["RTX" + i[1].split(" ")[1] + " " + i[1].split(" ")[2] + " " + i[1].split(" ")[3] if ("Ti" in i[1].split(" ")) and ("Max-Q" in i[1].split(" "))
                             else "RTX"  + i[1].split(" ")[1] + " " + i[1].split(" ")[2] + " " + i[1].split(" ")[3] if ("Super" in i[1].split(" ")) and ("Max-Q" in i[1].split(" "))
                             else "RTX"  + i[1].split(" ")[1] + " " + i[1].split(" ")[2] if "Ti" in i[1].split(" ")
                             else "RTX" + i[1].split(" ")[1] + " " + i[1].split(" ")[2] if "Super" in i[1].split(" ")
                             else "RTX"  + i[1].split(" ")[1] + " " + i[1].split(" ")[2] if "Max-Q" in i[1].split(" ")
                             else "RTX" + i[1].split(" ")[1] for i in q_df.gpu.str.split("RTX")]
    
    str_expr = "gpu.str.contains('gtx',case=False)" 
    q_df = df_copy.query(str_expr)
    q_index = q_df.index
    df_copy.loc[q_index,"gpu"] = ["GTX" + i[1].split(" ")[1] + " " + i[1].split(" ")[2] + " " + i[1].split(" ")[3] if ("Ti" in i[1].split(" ")) and ("Max-Q" in i[1].split(" "))
                             else "GTX" + i[1].split(" ")[1] + " " + i[1].split(" ")[2] + " " + i[1].split(" ")[3] if ("Super" in i[1].split(" ")) and ("Max-Q" in i[1].split(" "))
                             else "GTX" + i[1].split(" ")[1] + " " + i[1].split(" ")[2] if "Ti" in i[1].split(" ")
                             else "GTX" + i[1].split(" ")[1] + " " + i[1].split(" ")[2] if "Super" in i[1].split(" ")
                             else "GTX" + i[1].split(" ")[1] + " " + i[1].split(" ")[2] if "Max-Q" in i[1].split(" ")
                             else "GTX" + i[1].split(" ")[1] for i in q_df.gpu.str.split("GTX")]
    
    str_expr = "gpu.str.contains('amd',case=False)" 
    q_df = df_copy.query(str_expr)
    q_index = q_df.index
    df_copy.loc[q_index,"gpu"] = ["RX" + i[-1][1:] if i[-1] in [' 6600M',' 6700M',' 6800M',' 6800S',' 5500M',' 5600M',' 5300M',' 640',' 540',' 560X (Laptop)'] 
                             else "Radeon" + i[-1].split("Radeon")[-1][1:] if i[-1].split("Radeon")[-1][1:] in ['Pro 5500M','530','520','Pro Vega 16','Pro Vega 20']
                             else "integrated graphics" for i in q_df.gpu.str.split("RX")]
    df_copy.cpu = [i[-1] for i in df_copy.cpu.str.split(" ")]
    
    str_expr = "gpu.str.contains('geforce',case=False) or gpu.str.contains('Quadro',case=False) " 
    q_df = df_copy.query(str_expr)
    q_index = q_df.index
    df_copy.loc[q_index,"gpu"] = [i[-1].split("Quadro")[-1][1:].split(" ")[0] if "(Laptop)" in i[-1].split("Quadro")[-1][1:].split(" ")
                                  else i[-1].split("Quadro")[-1][1:] if "NVIDIA" in i[-1].split("Quadro")[0]
                                  else i[-1][1:] 
                                  for i in q_df.gpu.str.split("GeForce")]
    return df_copy
```
